The classic story *The Richest Man in Babylon* showed us a hierarchy of wealth building. Its key lessons are three levels of a investment pyramid: 

+ Level one: "save 10% of your salary"

  Survive and stabilize by keeping part of what you earn as the foundation of investing.
  
+ Level two: "seek expert's advice"
  
  Grow your savings **steadily** by seeking advice from who are competent. 
  
+ Level three: "make gold work for you"
  
  Accelerate the growth by reinvesting the children and grandchildren of your savings to grow a golden army.

Following the hierarchy of wealth building is crucial for long-term financial success because it prioritizes stability, minimizes catastrophic risks, and optimizes compound growth not only in **wealth itself**, but also the **options** of acquiring wealth. 

In this article, we want to design a hierarchy for lifetime wealth building, which aligns different sets of options with different wealth levels, balancing safety, growth, and risk tolerance at each stage. This follows a phased approach, starting from financial safety and progressing to high-scale proactive wealth acquisition. It is composed of the levels below:

+ Level 1 - Survive

  Goal: Save money

  Grow total asset to cover at least 3 months' cost of living through jobs or small businesses, then move up to investing the extra wealth in the certificate of deposit (CD, shortest term 3 months).

+ Level 2 - Stabilize savings 

  Goal: Save money and protect the capital from inflation 

  Grow total asset to cover at least 5 years' basics through jobs, businesses and passive income from CDs (or other low risk, high liquidity options), then move up to level 3. 

+ Level 3 - Grow passive income

  Goal: Grow passive income to cover the cost of living.

  Grow passive income through assets with moderate risk like broad index fund, real estate, dividend ETFs (Exchange-Traded Fund) till it can completely cover the cost of living.

+ Level 4 - Accelerate passive income

  Goal: Diversify into high-growth assets

  Add high-growth assets, such as growth stocks, dividend stocks, into the portfolio to produce more cash flow and appreciation. The cash flow can then be invested into other ventures, such as farm land, businesses, etc..

+ Level 5 - Proactive income

  Goal: Expand to proactive wealth building

  Add high-impact and high-control ventures, such as a micro-lending institution, a small bank, a credit card company, an insurance company, and a paid course or certification program for wealth building, to the strategy, in order to diversify from passive income to high-scale proactive profit.

The hierarchy can be visualized below:

```python
import matplotlib.pyplot as plt
import numpy as np

# set the style to a dark theme
plt.style.use("dark_background")

# match website background
plt.rcParams["figure.facecolor"] = "#181818"
plt.rcParams["axes.facecolor"] = "#181818"
plt.rcParams["axes.edgecolor"] = "#181818"


# Create figure
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_aspect('equal')
ax.axis('off')  # Hide axes

# Pyramid layers (from bottom to top)
layers = [
    {"label": "Survive", "width": 4, "color": "#4CAF50"},
    {"label": "Stabilize savings", "width": 3, "color": "#8BC34A"},
    {"label": "Grow passive income", "width": 2, "color": "#FFC107"},
    {"label": "Accelerate \npassive income", "width": 1.5, "color": "#FF9800"},
    {"label": "Scale up to \nproactive \nincome", "width": 1, "color": "#F44336"}
]

# Draw pyramid
y = 0
for i, layer in enumerate(layers):
    width = layer["width"]
    ax.barh(y, width, height=0.6, color=layer["color"], edgecolor='black')
    ax.text(0, y, f"{i+1}. {layer['label']}", 
            ha='left', va='center', color='black', fontweight='bold',
            fontsize=10)
    y += 0.6

# Title and annotations
plt.title("Investment Hierarchy Pyramid", fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```

## Conclusion ##

The hierarchy of wealth building shows a path from earning salary to financial freedom and to high-scale profit. It is also a path from a worker, who sells their labor and time to cover life expenses, to someone, who proactively acquires wealth through their position. 

Moving forward, the work done in this article also provides a framework for developing actual investing strategies to progress on the hierarchy.
