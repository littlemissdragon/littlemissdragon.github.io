The utility of investing in land is calculated as following:

$$
Utility = \frac{\text{Total return}}{\text{Total cost}}
$$

The total cost includes:
+ Cost of the land
+ Land assessment: such as land marking, soil tests, environmental checks
+ Closing cost: title transfer, legal work, etc.
+ Transportation and accommodation: if the land is located somewhere far away from where you live
+ Development: such as fencing and clearing
+ Ongoing taxes

The total return includes:
+ Land market price appreciation
+ Other potential returns, such as:
  + Leasing for crop farming, camping sites, etc..
  + Timber harvesting
